To install additional packages inside your `sample_fs.tbz2`, you will need to follow these steps:

1. **Extract the Tarball**: First, extract the contents of the `sample_fs.tbz2` file.
2. **Install the Packages**: Chroot into the extracted filesystem and install the necessary packages.
3. **Repack the Filesystem**: After installing the packages, repack the filesystem into a tarball.

Here's a detailed guide for each step:

### Step 1: Extract the Tarball

```bash
# Create a directory to extract the tarball
mkdir sample_fs
tar -xvjf sample_fs.tbz2 -C sample_fs
```

### Step 2: Install the Packages

You'll need to chroot into the extracted filesystem to install packages. Assuming you're working on a Linux system:

```bash
# Change root to the extracted filesystem
sudo chroot sample_fs /bin/bash

# Update package lists
apt-get update

# Install the packages you need (example packages: vim, git)
apt-get install vim git

# Exit chroot environment
exit
```

### Step 3: Repack the Filesystem

```bash
# Navigate out of the extracted directory
cd ..

# Create a new tarball
tar -cvjf sample_fs_new.tbz2 -C sample_fs .
```

Now you should have a new tarball `sample_fs_new.tbz2` that includes the additional packages you installed.

### Additional Notes

- **Network Access**: Ensure that your chroot environment has network access if you need to download packages from the internet.
- **Dependencies**: If you need specific versions of packages or additional dependencies, make sure to handle them appropriately within the chroot environment.
- **Disk Space**: Ensure you have enough disk space for both the extraction and repacking processes.
